```python
import pyautogui
import pyperclip
import pandas as pd


def openBrowser(browser='microsoft edge'):
    # Open Browser
    pyautogui.press('Win')
    pyperclip.copy(browser)
    pyautogui.hotkey('Ctrl', 'v')
    pyautogui.press('enter')


def newTabLink(link='https://google.com.br'):
    # New Tab
    pyautogui.hotkey('Ctrl', 't')
    pyperclip.copy(link)
    pyautogui.hotkey('Ctrl', 'v')
    pyautogui.press('enter')


def dlGDriveFile():
    # Download file
    pyautogui.sleep(2)
    pyautogui.click(437, 282, 1)    # Select File
    pyautogui.click(1769, 186, 1)   # Click on the 3 dots
    pyautogui.click(1543, 590, 1)   # Click on Download
    pyautogui.sleep(5)
    # After finishing the download, focus on browser
    pyautogui.click(951, 554, 1)


def sendEmail(email, subject='Test', msg='Hello via PyAutoGUI'):
    # Send Email via Gmail
    pyautogui.sleep(2)
    pyautogui.click(130, 198, 1)    # Select Email field
    pyperclip.copy(email)
    pyautogui.hotkey('Ctrl', 'v')

    pyautogui.press('Tab')          # Select Subject field
    pyperclip.copy(subject)
    pyautogui.hotkey('Ctrl', 'v')

    pyautogui.press('Tab')          # Select Message field
    pyperclip.copy(msg)
    pyautogui.hotkey('Ctrl', 'v')
    pyautogui.hotkey('Ctrl', 'Enter')


if __name__ == '__main__':

    pyautogui.PAUSE = 1

    browser = 'microsoft edge'
    openBrowser(browser)

    link_gdrive = 'https://drive.google.com/drive/folders/14oLE59U1RqyRqlBbKpsyymW-mitvbtoh?usp=sharing'
    newTabLink(link_gdrive)

    dlGDriveFile()

    linkEmail = 'https://mail.google.com/'
    newTabLink(linkEmail)

    file_location = '~/Downloads/Vendas - Dez.xlsx'
    table = pd.read_excel(r'~/Downloads/Vendas - Dez.xlsx')

    display(table)
    billing = table['Valor Final'].sum()
    sell_amount = table['Quantidade'].sum()
    print("Billing: ", billing, "Amount: ", sell_amount)

    email = 'plinio2xd+pythonimpressionador@gmail.com'
    subject = 'Relatório de vendas'
    msg = f"""
    Prezados, bom dia, tarde ou noite
    ---------- ---------- ---------- ---------- ---------- ----------
    O faturamento foi de R${billing:,.2f}
    A quantidade de produtos vendidos foi de {sell_amount:,} unidades
    ---------- ---------- ---------- ---------- ---------- ----------
    Abs
    Plínio, enviado de Arch Linux ~> VSCode ~> extensão Jupyter
    """
    sendEmail(email, subject, msg)
    pyautogui.sleep(3)

    pyautogui.hotkey('Ctrl', 'Shift', 'w')
```
